```python
class A:
  def __init__(self):
    self.P=10
    self.__Q=20
  def getY(self):
    return self.__Q
a=A()
print(a.P)
print(a.__Q)
```

The above error is because the \_\_Q variable is considered as a private variable, which cannot be accessed outside the class in this manner.
The variable \_\_Q can be accessed either through a.getY() method or a.\_A\_\_Q


Note: technically there is no private variables in python. Usually the \_\_variable is masked under the name for '\_className\_\_Variable' for avoiding clashes in inheritance chain, so the variables can still be accessed just not through the normal method that is "object.\_\_variable"

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def get_radius(self):
        return self.radius

    def get_area(self):
        return 3.14*self.radius**2
```

```python
c1 = Circle(5)
radius = c1.get_radius()
print("Radius of the circle:", radius)
area = c1.get_area()
print("Area of the circle:", area)
```

```python
class Student():
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

    def display_info(self):
        print("Name : ", self.name)
        print("Roll number : ", self.roll)

class AmritaStudent(Student):
    def __init__(self, name, roll, year):
        super().__init__(name, roll)
        self.year = year

    def display_info(self):
        super().display_info()
        print("Year:", self.year)

    def calculate_grade(self, marks):
        if marks >= 90:
            return 'A'
        elif marks >= 80:
            return 'B'
        elif marks >= 70:
            return 'C'
        elif marks >= 60:
            return 'D'
        else:
            return 'F'
```

```python
s1 = AmritaStudent("fufu", 88, 2022)
s1.display_info()
print("Grade : ", s1.calculate_grade(95))
```

```python
class Shape:
    def __init__(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, breadth):
        super().__init__()
        self.length = length
        self.breadth = breadth

    def calculate_area(self):
        return self.length * self.breadth

class Triangle(Shape):
    def __init__(self, base, height):
        super().__init__()
        self.base = base
        self.height = height

    def calculate_area(self):
        return 0.5 * self.base * self.height
```

```python
rectangle = Rectangle(7, 5)
triangle = Triangle(10, 5)
print("Area of rectangle:", rectangle.calculate_area())
print("Area of triangle:", triangle.calculate_area())
```

```python
# Base class1
class Mother:
    mothername = ""
    def mother(self):
        print(self.mothername)

# Base class2
class Father:
    fathername = ""
    def father(self):
        print(self.fathername)

# Derived class
class Son(Mother, Father):
    def parents(self):
        print("Father :", self.fathername)
        print("Mother :", self.mothername)

#Driver's code
s1 = Son()
s1.fathername = "RAM"
s1.mothername = "SITA"
s1. parents()
```

```python
class Parent:
    def __init__(self, name):
        self.name = name

    def getName(self):
        return self.name

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

    def getAge(self):
        return self.age

class Grandchild(Child):
    def __init__(self, name, age, location):
        super().__init__(name, age)
        self.location = location

    def getLocation(self):
        return self.location

gc = Grandchild("Srinivas", 24, "Hyderabad")
print(gc.getName(), gc.getAge(), gc.getLocation())
```
